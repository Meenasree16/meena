```python
import pymupdf as fitz  # PyMuPDF is now imported as fitz for compatibility

def extract_text_from_pdf(pdf_path):
    """
   Extracts text from each page of the PDF.
   Returns a list of dictionaries with page number and text.
    """
    doc = fitz.open(pdf_path)
    all_pages = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        text = page.get_text()
        all_pages.append({
            "page_num": page_num + 1,
            "text": text
        })

    return all_pages
```

```python
from sentence_transformers import SentenceTransformer

# Load a lightweight, high-quality model
embedder = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def flatten_chunks(scheme_chunks):
    """
    Converts each scheme's sections into a list of entries.
    Each entry has: scheme name, section name, text, page.
    """
    flat_sections = []

    for scheme in scheme_chunks:
        for section, content in scheme["sections"].items():
            flat_sections.append({
                "scheme_name": scheme["scheme_name"],
                "section_name": section,
                "text": content.strip(),
                "page_num": scheme.get("page_num", None)
            })

    return flat_sections
```

```python
# Flatten the schemes into one section per row
section_entries = flatten_chunks(chunks)

# Get only the text content
texts = [entry["text"] for entry in section_entries]

# Generate embeddings
embeddings = embedder.encode(texts, show_progress_bar=True)
```

```python
# Combine embedding and metadata
embedded_data = []

for i, entry in enumerate(section_entries):
    embedded_data.append({
        "embedding": embeddings[i],
        "scheme_name": entry["scheme_name"],
        "section_name": entry["section_name"],
        "text": entry["text"],
        "page_num": entry["page_num"]
    })
```

```python
import faiss
import numpy as np
```

```python
# Convert embeddings to NumPy array
embedding_matrix = np.array([entry["embedding"] for entry in embedded_data]).astype("float32")

# Create a FAISS index (L2 = cosine distance approximation)
index = faiss.IndexFlatL2(embedding_matrix.shape[1])
index.add(embedding_matrix)

print(f"✅ FAISS index built with {index.ntotal} chunks.")
```

```python
def retrieve_chunks(query, embedded_data, index, embedder, k=5):
    """
    Returns top-k most relevant chunks for the given query.
    """
    # Convert query to embedding
    query_embedding = embedder.encode([query]).astype("float32")

    # Search in FAISS
    distances, indices = index.search(query_embedding, k)

    # Get matching chunks
    results = [embedded_data[i] for i in indices[0]]

    return results
```

```python
query = "What is the subsidy percentage under PMEGP?"
top_chunks = retrieve_chunks(query, embedded_data, index, embedder, k=3)

for i, chunk in enumerate(top_chunks):
    print(f"\n🔹 Rank {i+1}: {chunk['scheme_name']} → {chunk['section_name']}")
    print(chunk["text"][:500])
```

```python
def retrieve_by_scheme(query, scheme_name_filter, embedded_data, index, embedder, k=5):
    # Step 1: Run full retrieval
    all_results = retrieve_chunks(query, embedded_data, index, embedder, k=20)

    # Step 2: Filter by scheme name
    filtered = [res for res in all_results if scheme_name_filter.lower() in res["scheme_name"].lower()]

    return filtered[:k]
```

```python
import torch

def generate_answer(query, retrieved_chunks, max_tokens=150):
    """
    Generates an answer from the TinyLLaMA model using RAG-style prompting.
    """
    # Combine all top retrieved contexts
    context = "\n\n".join([
        f"{chunk['section_name']}:\n{chunk['text']}" for chunk in retrieved_chunks
    ])

    # Prompt format: RAG-style
    prompt = f"""Answer the following question based only on the provided context.
Answer concisely and accurately.

Context:
{context}

Question:
{query}

Answer:"""

    # Tokenize
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            do_sample=False,
            temperature=0.7
        )

    # Decode and clean up
    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return output_text.split("Answer:")[-1].strip()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load your fine-tuned TinyLLaMA model
model_name = "Vipplav/tinyllama-finetuned-faq"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# --- 10. Interactive Inference ---

print("\n🔍 Ready! Ask questions based on Handbook.pdf")
print("Type 'exit' to quit.")

def rag_query(user_query):
    # Retrieve top relevant chunks (top 3)
    retrieved = retrieve_chunks(user_query, embedded_data, index, embedder, k=3)

    # OR use filtered retrieval by scheme if desired:
    # retrieved = retrieve_by_scheme(user_query, "PMEGP", embedded_data, index, embedder, k=3)

    # Generate answer using model
    answer = generate_answer(user_query, retrieved)
    return answer

# CLI loop
while True:
    user_query = input("You: ")
    if user_query.strip().lower() == "exit":
        break

    answer = rag_query(user_query)
    print(f"Bot: {answer}\n")
```
